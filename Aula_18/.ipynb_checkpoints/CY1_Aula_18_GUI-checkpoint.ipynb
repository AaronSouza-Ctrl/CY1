```python
import CY1_Aula_18_ChatbotBase as pc
from tkinter import *

main_window = Tk()
main_window.title('Dona Maria')
main_window.geometry('1000x700')
main_window.grid()

fontePadrao = ('Arial', '12')
fonte14 = ('Arial', '14')

frame = Frame(main_window)
frame.grid()

l_identify = Label(frame, text = 'Insira sua mensagem aqui: ', font = fonte14)
l_identify.grid(row = 0, column = 0)

e_mensagem = Entry(frame, font = fontePadrao)
e_mensagem.grid(row = 0, column = 1)

frame2 = Frame(main_window)
frame2.grid(row = 1, column = 0)
v = StringVar()
Label(frame2, textvariable = v, font = fontePadrao).grid()

nome_maquina = 'Maria'
v.set('Qual é o seu nome? ')
entrada_sugestao = False
entrada_nome_usuario = True
nome_usuario = ''

def roda_Chatbot():
    global entrada_sugestao
    global entrada_nome_usuario
    global historico_conversa
    global nome_usuario

    if entrada_nome_usuario:
        nome_usuario = e_mensagem.get()
        saudacao = pc.saudacao_GUI(nome_maquina, nome_usuario)
        historico_conversa = '{}: {}\n'.format(nome_maquina, saudacao)
        v.set(historico_conversa)
        entrada_nome_usuario = False

    else:
        texto = e_mensagem.get()
        historico_conversa += '\n{}: {}'.format(nome_usuario, texto)
        v.set(historico_conversa)

        if entrada_sugestao:
            pc.salva_sugestao(texto)
            entrada_sugestao = False
            historico_conversa += '\n Agora aprendi! Vamos continuar nossa conversa... \n'
            v.set(historico_conversa)
        else:
            resposta = pc.buscaResposta_GUI('Cliente: {}\n'.format(texto))
            if resposta == 'Não vem não! Me respeite!':
                historico_conversa += '\n{}: Não vem não! Me respeite!\n'.format(nome_maquina)
                v.set(historico_conversa)
                texto = e_mensagem.get()
                resposta = pc.buscaResposta_GUI('Cliente: {}\n'.format(texto))

            elif resposta == 'Ok. Você está fazendo um (a) reclamação, elogio ou sugestão':
                historico_conversa += '\n{}: Ok. Você está fazendo um (a) reclamação, elogio ou sugestão\n'.format(nome_maquina)
                v.set(historico_conversa)
                texto = e_mensagem.get()
                resposta = pc.buscaResposta_GUI('Cliente: {}\n'.format(texto))

            else:
                if resposta == 'Me desculpe, não sei o que falar':
                    historico_conversa += '\n Me desculpe, não sei o que falar. O que você esperava? \n'
                    v.set(historico_conversa)
                    entrada_sugestao = True
                else:
                    historico_conversa += '\n' + pc.exibeResposta_GUI(texto, resposta, nome_maquina)
                    v.set(historico_conversa)

Button(frame, text = 'Clique', command = roda_Chatbot, font = fontePadrao).grid(row = 0, column = 2)

main_window.mainloop()
```
