```python
import requests as r
from bs4 import BeautifulSoup

try:
    resultado = r.get('https://www.globo.com')

except Exception as erro:
    print('X_X', erro)

else:
    resposta = resultado.content
    soup = BeautifulSoup(resposta, 'html.parser')
    manchete = soup.find('h1')
    if manchete != None:
        print(manchete.text)
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.uol.com.br/start/esport/'
page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')

lista = ['lol', 'valorant', 'fifa', 'fortnite', 'free fire', 'among us', 'minecraft']

for paragrafo in soup.find_all('body'):
    for palavra in lista:
        for paragrafo_str in paragrafo.stripped_strings:
            if palavra.upper() in str(paragrafo_str).upper():
                print('NOTÍCIA SOBRE: {}\n{}\n'.format(palavra.upper(), paragrafo_str))
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.uol.com.br'
page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')
links = soup.find_all('a')
links_dict = {}

for link in links:
    if (link.h3 != None) and (link.h3.get('class') != None):
        links_dict[link.h3.text] = link.get('href')

for k, v in links_dict.items():
    print('{}: {}'.format(k, v))

df = pd.DataFrame({
    'Notícia': links_dict.keys(),
    'Link': links_dict.values()
    })

df.head()
df.to_excel('links_uol.xlsx')
```
